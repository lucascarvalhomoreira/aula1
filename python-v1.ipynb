## 6. Loops 
Loops em uma linguagem de programação são usados para executar um determinado conjunto de instruções de forma iterativa. Python nos dá basicamente três maneiras diferentes pelas quais os loops podem ser implementados dentro do programa.
* Loop While
* Loop For

### 6.1 Loop While

Os loops while são usados para iterar sobre um determinado conjunto de instruções várias vezes até que a condição de teste se torne falsa. É mais adequado usar esse loop quando não sabemos quantas iterações queremos executar no loop. É por isso que os loops while se enquadram na categoria de iteração indefinida.

```python
a = 1 # a recebe 1
while a<=10:
    print(a)
    a = a + 1
print('Acabou!!')
```

Mais formalmente, aqui está o fluxo de execução para uma instrução while:

1. Determine se a condição é verdadeira ou falsa.

2. Se for falsa, saia da instrução while e continue a execução da próxima instrução.

3. Se a condição for verdadeira, execute o corpo e então volte ao passo 1.

Este tipo de fluxo chama-se loop (**laço**), porque o terceiro passo faz um loop de volta ao topo.

***Exemplos***

* Contagem regressiva

```python
a = 10
while a>0:
    print(a)
    a = a - 1
print('Fim')
```

* Percorrendo os elementos de uma lista. 

```python
lista = ['a','e','i','o','u']

i = 0
while i<5:
    print(lista[i])
    i = i + 1
```

```python
lista = ['a','e','i','o','u']
# a função len(lista)retorna o tamanho a lista
tamanho_lista = len(lista) 
i = 0

while i<tamanho_lista:
    print(lista[i])
    i = i + 1
```

```python
lista = ['a','e','i','o','u']

i = 0
while i<len(lista):
    print(lista[i])
    i = i + 1
```

```python
lista = ['a','e','i','o','u']

print(lista[0])
print(lista[2])
print(lista[4])
```

### 6.2 Loop For

O loop for é uma das maneiras mais convenientes de iterar sobre um tipo de sequência ou outro objeto iterável em Python. Essa sequência pode ser List, String, Tuple, Dictionary, Set, etc. Usando o loop for, podemos iterar em uma sequência até chegarmos ao final da sequência. Assim que entramos no final da sequência, saímos do loop. A iteração sobre um tipo de sequência também é conhecida como uma travessia da sequência. Uma de suas vantagens sobre o loop while é que ele não requer que uma variável de indexação seja definida no início.

```python
for i in range(0,10):
    print(i)
```

Com a função range() , podemos definir um step (um passo), que é o intervalo entre os elementos.
Por padrão, o step tem valor igual a 1, mas podemos alterar este valor passando um terceiro parâmetro 

```python
for i in range(0,10,2):
    print(i)
```

```python
for i in range(10,0,-1):
    print(i)
```

```python
a = 2
b = 7
for i in range(a,b):
    print(i)
```

```python
for letra in 'Olá Mundo':
    print(letra)
```

```python
for elemento in ['Jose','Maria','João']:
    print(elemento)
```

**Exemplos**

* Repetindo de 1 até um valor lido fo teclado.

```python
valor = input('Digite um valor: ')

valor_inteiro = int(valor)  # transformar a string valor em um inteiro

for i in range(1,valor_inteiro):
    print(i)
```

## 7. Dicionário

Um dicionário é uma estrutura de dados do tipo coleção. É, portanto, um objeto que contém mais que um valor. 

Se nas listas, os elementos são acessados por meio de uma posição ou índice, nos dicionários isso ocorre de forma diferente. O acesso às informações acontece por meio de chaves. 

A sintaxe para criação de dicionários em Python é:

**{chave1 : valor1, chave2 : valor2, …​}**.

Assim como um dicionário de chaves (termos) que estão associadas a valores (significados dos termos), os dicionários em Python são uma estrutura de dados que nos permite mapear chaves a valores.

### 7.1 Como criar dicionário Python

Para criar um dicionário em Python, basicamente, as informações precisam ser passadas entre chaves {}, que por sua vez precisam seguir a estrutura de pares **chave:valor**  ou **key:value**. 

```python
alunos = { 
    '123456':'José', 
    '654321':'Maria'}

print(alunos)

print(alunos['123456'])  # para consultar um elemento do dicionário
print(alunos['654321'])
```

```python
alunos['321456'] = 'Rita' # para inserir um novo elemento
print(alunos)
```

```python
alunos['894091'] = 'João'
print(alunos)
print(len(alunos))
```

```python
x = alunos.pop('321456') # a função permite remover um elemento
print("Elemento",x)
print(alunos)
```

```python
x = alunos['894091']  # somente consulta sem remover o elemento
print("Elemento",x)
print(alunos)
```

Recomenda-se que o nome do dicionário seja o mais explícito possível. Por exemplo, se tivermos um dicionário de pessoas mapeadas para as cidades em que vivem, um dicionário pode ser criado da seguinte maneira:


```python
ra_alunos = { 
    '123456':'José', 
    '654321':'Maria'}
```

```python
porta_protocolo = {
    '22':'ssh',
    '80':'http',
    '20':'ftp'
}

print("Procololo porta 22", porta_protocolo['22'])
```

Os elementos de um dicionário são armazenados de forma não ordenada.

Isso ocorre de forma que seus elementos sejam exclusivos. Porém, os elementos de um mesmo valor podem ser duplicados, uma vez que a chave é diferente. 

A estrutura se assemelha à tabela de um banco de dados quando se tem uma chave primária composta.

```python
ra_alunos = { 
    '123456':'José', 
    '654321':'Maria'}

print(ra_alunos)

ra_alunos['123456'] = 'João'
print(ra_alunos)
```

Para juntar dois dicionários, é possível usar a função **update([other])** , como indicado na documentação do Python.

```python
outro_dicionario = {}   # dicionário vazio
outro_dicionario['782234'] = 'Ana'
outro_dicionario['746363'] = 'Paulo'

print("Dicionário outro_dicionario > ",outro_dicionario)
print("Dicionário ra_alunos > ", ra_alunos)

ra_alunos.update(outro_dicionario)
print("Dicionário ra_alunos > ", ra_alunos)
```

# 8. Funções 

Uma das principais formas de se criar componentes reusáveis em Python (e em outras linguagens de programação também) são as chamadas **funções**. 

Uma **função** é basicamente um trecho de código (um procedimento, um algoritmo) que pode ser usado várias vezes.  Quando surge essa necessidade, para que as instruções de um trecho do programa não precisem ser repetidas em diferentes partes de um programa, elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada quando for necessária.

Apesar de só estarmos falando sobre funções agora, nós já usamos funções várias vezes nas aulas anteriores. Por exemplo, quando fazemos **print(len(alunos))** estamos usando a função **len()** da biblioteca padrão Python para determinar o comprimento da estrutra *alunos* Agora aprenderemos a criar e usar nossas próprias funções em Python.

## 8.1 Criando funções no Python

A sintaxe de uma função é definida por três partes: nome, parâmetros e corpo, o qual agrupa uma sequência de linhas que representa algum comportamento. 

No código a seguir, temos um exemplo de declaração de função em Python:

```python
def hello():
    print("Olá!!")
```

Essa função, de nome **hello**, tem como objetivo imprimir a string *Olá*.

A palavra reservada **def**, na primeira linha, explicita a definição da função naquele ponto. Em seguida, entre parênteses, teriamos os parâmetro, nesse caso não existem parânetros em nossa função. Ainda na mesma linha, observe a utilização dos dois pontos (:), que indicam que o código identado nas linhas abaixo faz parte da função que está sendo criada. 

É importante ressaltar que, para respeitar a sintaxe da linguagem, a linha 2 está avançada em relação à linha 1.

Para executar a função, de forma semelhante ao que ocorre em outras linguagens, devemos simplesmente chamar seu nome e passar os parâmetros esperados entre parênteses.

```python
hello()
```

Nessa nova implementação da função **hello** temos o parâmetro **meu_nome**. 

```python
def hello(nome_usuario):
    print("Olá ", nome_usuario,"!")
```

## 8.2 A instrução return 

Uma função pode retornar um valor ao final de sua execução. Para isso devemos utilizar o comando **return**. 

A função **soma**, definida a seguir, tem com parâmetros de entrada dois valores inteiro e ela retorna o valor obtido com a soma de seus parâmetros.

```python
def soma(n1,n2):
    resultado = n1 + n2
    return resultado
```

Para usarmos o valor calculado na função soma devemos armazena-lo em uma varável que irá receber o valor retornado:

```python
valor_soma =  soma(10,15) 
print("o valor é igual a: ",valor_soma)
```

```python
print("o valor é igual a: ",soma(1,6))
```

```python
def resto_divisao(dividendo,divisor):
    '''
        Calcula e retorna o resto de uma divisão.
    '''
    resto = dividendo %  divisor
    return resto
    
    

x = resto_divisao(12,5)
print(x)

print(resto_divisao(12,3))
```

```python
def eh_par(numero):
    '''
        Verifica se o numero é par, se for retorna true e false caso
        contrário.
    '''
    resto = numero % 2 # se o numero for par o resto da divisão por 2 é 0
    if resto==0:
        return True
    else:
        return False

print("o número 5 é par? ", eh_par(5)) 
print("o número 8 é par? ",eh_par(8))
```

```python
# Programa que le um valor do teclado e determina se esse é par ou impar.
valor = input("Digite um valor: ")
valor_int = int(valor)   # transforma o valor lido em um inteiro

if eh_par(valor_int):
    print("Esse valor é par")
else:
    print("Esse valor é impar")
```

# 9. Módulos em Python

A modularização de uma aplicação é algo extremamente importante. Ela deve ser feita por duas razões:
* Para escrever módulos e reutilizar o código escrito em outras aplicações;
* Queremos que o código de nossa aplicação esteja organizado de modo a facilitar o controle e o entendimento do todo, facilitando assim a expansão organizada e a manutenção.

Em poucas palavras: **módulos** são arquivos de código Python cuja interface é conhecida e que podem ser importados por outros módulos. 

Dizer que a "interface é conhecida" significa que quando um programador importar um módulo, ele saberá (ou tem meios de saber) quais funções e classes o módulo possui. Ele também saberá como usá-las, isto é, conhecendo seus nomes, parâmetros, que exceções pretende tratar, dentre outras características.


Na linguagem Python, os módulos são nada mais que códigos-fonte que podemos importar em nossos scripts. Estes arquivos podem ter qualquer estrutura do Python e são executados no momento que são importados. Caso seja necessário executar o módulo novamente durante a execução do programa, podemos utilizar a função reload().

## 9.1 Importando módulos

### Pacotes

Um pacote é uma coleção de módulos.

Quando nossos módulos aumentam de tamanho, podemos dividí-los em pacotes, segundo o tutorial do Python:

*“O uso de pacotes permite aos autores de grupos de módulos (como NumPy ou PIL) não terem que se preocupar com colisão entre os nomes de seus módulos e os nomes de módulos de outros autores.”*

Ainda na documentação:

*“Pacotes são uma maneira de estruturar espaços de nomes para módulos Python utilizando a sintaxe de **‘nomes pontuados’** (dotted names)”.*

```python
import math

x = math.sqrt(36)  # funçao da math que calcula a raiz quadrada
print("A raiz quadrada de 36 é: ",x)
print(math.pi)     # o valor de PI que está definido na biblioteca math

coseno_x = math.cos(math.pi)
print("valor do coseno de Pi é:",coseno_x )
```

### json

**JSON** (*JavaScript Object Notation*) é um formato de dados popular usado para representar dados estruturados. É comum transmitir e receber dados entre um servidor e uma aplicação web no formato JSON.

Para manipular estruturas em JSON (strings ou arquivos contendo objetos em JSON), você pode usar o módulo **json** do Python. Você precisa importar o módulo antes de poder usá-lo.

```python
# essa linha importa os módulos
import requests
import json

cep = "14030000"
url = "http://viacep.com.br/ws/" + cep + "/json/"

print("URL: ",url)

resposta = requests.get(url)

if resposta.status_code == 200:
    dados = resposta.content
    
    dados_json = json.loads(dados)
    
    print(dados_json)
    print(dados_json['cep'])
    print(dados_json['localidade'])
else:
    print("Erro: ",resposta.status_code)
```

A função **loads** em Python faz parte do módulo json e é usada para carregar (ou decodificar) uma string JSON em um objeto Python. A sigla "loads" significa "load string" (carregar string).

O comando que faz o oposto, ou seja, converte um dicionário do Python em uma string no formato JSON é a função **json.dumps()**. A sigla "dumps" significa "dump string" (descarregar em forma de string).

Ao chamar json.dumps(dicionario), você está passando um dicionário Python como argumento para a função dumps. Essa função irá serializar o dicionário em uma string JSON.

Aqui está um exemplo de uso da função dumps:

```python
import json

dicionario = {"nome": "Maria", "idade": 25, "cidade": "Campinas"}
string_json = json.dumps(dicionario)

print(string_json) 
```

### socket

A programação de socket é uma maneira de conectar dois nodes em uma rede para se comunicarem. Um socket (node) escuta em uma determinada porta em um IP, enquanto o outro socket alcança o outro para formar uma conexão. O servidor forma o socket do ouvinte enquanto o cliente alcança o servidor.
Eles são a verdadeira espinha dorsal por trás da navegação na web. Em termos mais simples, existe um servidor e um cliente.
A programação de socket é iniciada importando o módulo **socket** e criando um socket simples.

```python
# Exemplo de script para se conectar ao Google usando socket
import socket 
 
try:
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    print ("Socket criado com sucesso!")
except socket.error as err:
    print ("Erro na criação do socket: erro %s" %(err))
 
# porta padrão para o  socket
porta = 80
 
try:
    # encontra o endereço IP de uma URL
    host_ip = socket.gethostbyname('www.google.com')
    # conecando com o servidor
    s.connect((host_ip, porta)) 
    print ("O socket foi conectado com sucesso ao google")
except socket.gaierror:
    # isso significa que não foi possível resolver o host
    print ("ocorreu um erro ao resolver o host")
```
